```python
import numpy as np

# 系統矩陣 A 和常數向量 b
A = np.array([
    [ 4, -1,  0, -1,  0,  0],
    [-1,  4, -1,  0, -1,  0],
    [ 0, -1,  4,  0,  1, -1],
    [-1,  0,  0,  4, -1, -1],
    [ 0, -1,  0, -1,  4, -1],
    [ 0,  0, -1,  0, -1,  4]
])

b = np.array([0, -1, 9, 4, 8, 6])

# 初始解
x0 = np.zeros_like(b)

# 誤差容許值與最大疊代次數
tol = 1e-6
max_iter = 1000

def jacobi(A, b, x0, tol, max_iter):
    n = len(b)
    x = x0.copy()
    for k in range(max_iter):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k+1
        x = x_new
    return x, max_iter

x_jacobi, it_jacobi = jacobi(A, b, x0, tol, max_iter)
print("Jacobi solution:", x_jacobi)
print("Iterations:", it_jacobi)

def gauss_seidel(A, b, x0, tol, max_iter):
    n = len(b)
    x = x0.copy()
    for k in range(max_iter):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k+1
        x = x_new
    return x, max_iter

x_gs, it_gs = gauss_seidel(A, b, x0, tol, max_iter)
print("Gauss-Seidel solution:", x_gs)
print("Iterations:", it_gs)

def sor(A, b, x0, omega, tol, max_iter):
    n = len(b)
    x = x0.copy()
    for k in range(max_iter):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (1 - omega) * x[i] + omega * (b[i] - s1 - s2) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k+1
        x = x_new
    return x, max_iter

omega = 1.25  # 可以微調
x_sor, it_sor = sor(A, b, x0, omega, tol, max_iter)
print("SOR solution:", x_sor)
print("Iterations:", it_sor)

def conjugate_gradient(A, b, x0, tol, max_iter):
    x = x0.copy()
    r = b - A @ x
    p = r.copy()
    rs_old = np.dot(r, r)

    for i in range(max_iter):
        Ap = A @ p
        alpha = rs_old / np.dot(p, Ap)
        x = x + alpha * p
        r = r - alpha * Ap
        rs_new = np.dot(r, r)
        if np.sqrt(rs_new) < tol:
            return x, i+1
        p = r + (rs_new / rs_old) * p
        rs_old = rs_new

    return x, max_iter

x_cg, it_cg = conjugate_gradient(A, b, x0, tol, max_iter)
print("Conjugate Gradient solution:", x_cg)
print("Iterations:", it_cg)
```
